### 3. [簡答題] 承上題，請問為什麼依照 index 合併會發生錯誤？請用程式解決。

````python
'''
因為columns名重覆,index會直接報錯

'''
#方法1
df1=df1.rename(columns={"fruit": "fruit_df1", "weight": "weight_df1","price": "price_df1"})
df2=df2.rename(columns={"fruit": "fruit_df2", "weight": "weight_df2","price": "price_df2"})
print(df1)
print(df2)
df1.join(df2)
````

````python
#方法二
df1.join(df2, lsuffix='_df1', rsuffix='_df2')
````
